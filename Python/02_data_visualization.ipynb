```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

print("="*80)
print("QUESTION 2: Data Visualization and Analysis from Excel File")
print("="*80)

# ============================================================================
# Part A: Read Excel File and Create Visualizations
# ============================================================================
print("\n" + "="*80)
print("PART A: Reading Excel File and Creating Visualizations")
print("="*80)

# Read the Excel file
excel_file_path = '../data/algorithm_times.xlsx'

if not os.path.exists(excel_file_path):
    print(f"\nError: File not found at {excel_file_path}")
    print("Please ensure the Excel file exists in the data/ directory")
else:
    # Read the Excel file
    df = pd.read_excel(excel_file_path)
    
    print(f"\nExcel file loaded successfully from: {excel_file_path}")
    print(f"\nOriginal DataFrame shape: {df.shape[0]} rows × {df.shape[1]} columns")
    print("\nOriginal DataFrame contents:")
    print(df)
    
    # ========================================================================
    # Part B: Add New Data Automatically
    # ========================================================================
    print("\n" + "="*80)
    print("PART B: Adding New Data (700KB, 80, 320, 700)")
    print("="*80)
    
    # Get column names
    col_names = df.columns.tolist()
    
    # Check if 700KB data already exists
    data_exists = False
    for idx, row in df.iterrows():
        first_col_val = str(row[col_names[0]])
        if '700' in first_col_val:
            data_exists = True
            break
    
    if data_exists:
        print("\n700KB data already exists in the Excel file. Skipping addition.")
    else:
        print("\n700KB data not found. Adding new row...")
        
        # Create new row
        new_row = pd.DataFrame({
            col_names[0]: ['700KB'],
            col_names[1]: [80],
            col_names[2]: [320],
            col_names[3]: [700]
        })
        
        # Add new row to dataframe
        df = pd.concat([df, new_row], ignore_index=True)
        
        # Save updated dataframe back to Excel
        df.to_excel(excel_file_path, index=False)
        
        print("✓ New data added successfully to Excel file!")
        print("\nUpdated DataFrame:")
        print(df)
    
    # ========================================================================
    # Extract data dynamically (after potential update)
    # ========================================================================
    data_sizes = df.iloc[:, 0].values  # First column (Data size)
    alg1_times = df.iloc[:, 1].values  # Second column (Algorithm 1)
    alg2_times = df.iloc[:, 2].values  # Third column (Algorithm 2)
    alg3_times = df.iloc[:, 3].values  # Fourth column (Algorithm 3)
    
    # Get column names dynamically
    data_size_label = col_names[0]
    alg1_label = col_names[1]
    alg2_label = col_names[2]
    alg3_label = col_names[3]
    
    print(f"\nFinal data being used for visualization:")
    print(f"  {data_size_label}: {data_sizes}")
    print(f"  {alg1_label}: {alg1_times}")
    print(f"  {alg2_label}: {alg2_times}")
    print(f"  {alg3_label}: {alg3_times}")
    
    # ========================================================================
    # Visualization 1: Horizontal Bar Chart (Grouped)
    # ========================================================================
    print("\n" + "-"*80)
    print("Creating Horizontal Bar Chart...")
    print("-"*80)
    
    fig, ax = plt.subplots(figsize=(12, 10))
    
    y_pos = np.arange(len(data_sizes))
    bar_height = 0.25
    
    bars1 = ax.barh(y_pos - bar_height, alg1_times, bar_height,
                     label=alg1_label, color='#264653', alpha=0.9, edgecolor='white')
    bars2 = ax.barh(y_pos, alg2_times, bar_height,
                     label=alg2_label, color='#2A9D8F', alpha=0.9, edgecolor='white')
    bars3 = ax.barh(y_pos + bar_height, alg3_times, bar_height,
                     label=alg3_label, color='#E76F51', alpha=0.9, edgecolor='white')
    
    # Add value labels at the end of bars
    for bars in [bars1, bars2, bars3]:
        for bar in bars:
            width = bar.get_width()
            ax.text(width, bar.get_y() + bar.get_height()/2,
                    f' {int(width)}',
                    ha='left', va='center', fontsize=9, fontweight='bold')
    
    ax.set_ylabel('Data Size', fontsize=14, fontweight='bold')
    ax.set_xlabel('Execution Time (ms)', fontsize=14, fontweight='bold')
    ax.set_title('Algorithm Execution Times - Horizontal Bar Comparison',
                 fontsize=16, fontweight='bold', pad=20)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(data_sizes, fontsize=11)
    ax.legend(fontsize=12, loc='lower right')
    ax.grid(axis='x', alpha=0.3, linestyle='--')
    ax.invert_yaxis()
    
    plt.tight_layout()
    plt.savefig('../outputs/figures/question2_bar_chart.png', dpi=300, bbox_inches='tight')
    print("Bar chart saved to: ../outputs/figures/question2_bar_chart.png")
    plt.show()
    
    # ========================================================================
    # Visualization 2: Filled Area / Line Chart
    # ========================================================================
    print("\n" + "-"*80)
    print("Creating Filled Area Chart...")
    print("-"*80)
    
    fig, ax = plt.subplots(figsize=(14, 8))
    
    x_pos = np.arange(len(data_sizes))
    
    ax.plot(x_pos, alg1_times, marker='o', linewidth=2.5, markersize=9,
            label=alg1_label, color='#264653')
    ax.fill_between(x_pos, alg1_times, alpha=0.15, color='#264653')
    
    ax.plot(x_pos, alg2_times, marker='s', linewidth=2.5, markersize=9,
            label=alg2_label, color='#2A9D8F')
    ax.fill_between(x_pos, alg2_times, alpha=0.15, color='#2A9D8F')
    
    ax.plot(x_pos, alg3_times, marker='^', linewidth=2.5, markersize=9,
            label=alg3_label, color='#E76F51')
    ax.fill_between(x_pos, alg3_times, alpha=0.15, color='#E76F51')
    
    # Add value labels on points
    for i in range(len(data_sizes)):
        ax.annotate(f'{int(alg1_times[i])}', (x_pos[i], alg1_times[i]),
                    textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9)
        ax.annotate(f'{int(alg2_times[i])}', (x_pos[i], alg2_times[i]),
                    textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9)
        ax.annotate(f'{int(alg3_times[i])}', (x_pos[i], alg3_times[i]),
                    textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9)
    
    ax.set_xlabel('Data Size', fontsize=14, fontweight='bold')
    ax.set_ylabel('Execution Time (ms)', fontsize=14, fontweight='bold')
    ax.set_title('Algorithm Execution Times - Filled Area Chart',
                 fontsize=16, fontweight='bold', pad=20)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(data_sizes, fontsize=11)
    ax.legend(fontsize=12, loc='upper left')
    ax.grid(True, alpha=0.3, linestyle='--')
    
    plt.tight_layout()
    plt.savefig('../outputs/figures/question2_line_chart.png', dpi=300, bbox_inches='tight')
    print("Line chart saved to: ../outputs/figures/question2_line_chart.png")
    plt.show()
    
    # ========================================================================
    # Visualization 3: Violin Plot
    # ========================================================================
    print("\n" + "-"*80)
    print("Creating Violin Plot...")
    print("-"*80)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Prepare data for violin plot
    violin_data = [alg1_times, alg2_times, alg3_times]
    violin_labels = [alg1_label, alg2_label, alg3_label]
    
    vp = ax.violinplot(violin_data, showmeans=True, showmedians=True, showextrema=True)
    
    # Color each violin differently
    colors = ['#264653', '#2A9D8F', '#E76F51']
    for patch, color in zip(vp['bodies'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.6)
        patch.set_edgecolor('black')
        patch.set_linewidth(1.5)
    
    vp['cmeans'].set_color('green')
    vp['cmeans'].set_linewidth(2.5)
    vp['cmedians'].set_color('red')
    vp['cmedians'].set_linewidth(2.5)
    
    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(violin_labels, fontsize=11)
    ax.set_xlabel('Algorithms', fontsize=14, fontweight='bold')
    ax.set_ylabel('Execution Time (ms)', fontsize=14, fontweight='bold')
    ax.set_title('Algorithm Execution Times - Violin Plot Distribution',
                 fontsize=16, fontweight='bold', pad=20)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Add legend for mean and median
    from matplotlib.lines import Line2D
    legend_elements = [Line2D([0], [0], color='red', linewidth=2.5, label='Median'),
                        Line2D([0], [0], color='green', linewidth=2.5, label='Mean')]
    ax.legend(handles=legend_elements, fontsize=11, loc='upper right')
    
    plt.tight_layout()
    plt.savefig('../outputs/figures/question2_box_plot.png', dpi=300, bbox_inches='tight')
    print("Violin plot saved to: ../outputs/figures/question2_box_plot.png")
    plt.show()
    
    # ========================================================================
    # Statistical Summary
    # ========================================================================
    print("\n" + "="*80)
    print("STATISTICAL SUMMARY")
    print("="*80)
    
    summary_stats = pd.DataFrame({
        'Algorithm': [alg1_label, alg2_label, alg3_label],
        'Mean': [np.mean(alg1_times), np.mean(alg2_times), np.mean(alg3_times)],
        'Median': [np.median(alg1_times), np.median(alg2_times), np.median(alg3_times)],
        'Std Dev': [np.std(alg1_times), np.std(alg2_times), np.std(alg3_times)],
        'Min': [np.min(alg1_times), np.min(alg2_times), np.min(alg3_times)],
        'Max': [np.max(alg1_times), np.max(alg2_times), np.max(alg3_times)]
    })
    
    print("\n")
    print(summary_stats.to_string(index=False))
    
    # ========================================================================
    # Part C: Calculate Average for Algorithm 2 (100-600KB data only)
    # ========================================================================
    print("\n" + "="*80)
    print("PART C: Average Execution Time for Algorithm 2")
    print("="*80)
    
    # Convert data_sizes to numeric format for comparison
    data_sizes_numeric = []
    for size in data_sizes:
        if isinstance(size, (int, float)):
            data_sizes_numeric.append(int(size))
        else:
            # Extract numeric value from string (e.g., "100KB" -> 100)
            numeric_str = ''.join(filter(str.isdigit, str(size)))
            data_sizes_numeric.append(int(numeric_str) if numeric_str else 0)
    
    # Get indices where data size is between 100 and 600
    indices_100_600 = [i for i, size in enumerate(data_sizes_numeric) if 100 <= size <= 600]
    
    # Extract Algorithm 2 times for 100-600KB range
    alg2_times_filtered = [alg2_times[i] for i in indices_100_600]
    
    print(f"\n{alg2_label} execution times for data sizes 100KB to 600KB:")
    for i in indices_100_600:
        print(f"  {data_sizes[i]}: {alg2_times[i]} ms")
    
    # Calculate average
    avg_alg2 = np.mean(alg2_times_filtered)
    
    print(f"\nAverage execution time for {alg2_label} (100KB-600KB): {avg_alg2:.2f} ms")
    
    # Save results to text file
    with open('../outputs/results/question2_results.txt', 'w', encoding='utf-8') as f:
        f.write("="*80 + "\n")
        f.write("QUESTION 2: ANALYSIS RESULTS\n")
        f.write("="*80 + "\n\n")
        
        f.write("PART A: Data Summary\n")
        f.write("-"*80 + "\n")
        f.write(df.to_string(index=False))
        f.write("\n\n")
        
        f.write("Statistical Summary:\n")
        f.write(summary_stats.to_string(index=False))
        f.write("\n\n")
        
        f.write("PART C: Algorithm 2 Average (100KB-600KB)\n")
        f.write("-"*80 + "\n")
        f.write(f"Average execution time: {avg_alg2:.2f} ms\n")
        f.write(f"Number of data points: {len(alg2_times_filtered)}\n")
        f.write(f"Data range: {data_sizes[indices_100_600[0]]} to {data_sizes[indices_100_600[-1]]}\n")
    
    print("\nResults saved to: ../outputs/results/question2_results.txt")
    
    print("\n" + "="*80)
    print("QUESTION 2 COMPLETED SUCCESSFULLY")
    print("="*80)
```
